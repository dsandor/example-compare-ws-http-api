```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display, HTML

websocket_df = pd.read_csv('websocket_performance_call_times_0ms_auth_delay.csv')
httpapi_df = pd.read_csv('httpapi_performance_call_times_0ms_auth_delay.csv')

websocket_df['type'] = 'WebSocket'
httpapi_df['type'] = 'HttpApi'

call_times_df = pd.concat((websocket_df, httpapi_df))

display(HTML(call_times_df.head().to_html()))
```

```python
import seaborn as sns

# Histogram
fig, ax = plt.subplots()
plt.hist(websocket_df['call_time'], bins=20)
plt.title('Histogram of WebSocket call times', fontsize=18)
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Frequency')
plt.show()
print("")

# Scatterplot
plt.title('Scatter plot WebSocket call times (ms)', fontsize=18)
sns.regplot(x="index", y="call_time", data=websocket_df.reset_index(), ci=None)
plt.show()
```

```python
import seaborn as sns

# Histogram
fig, ax = plt.subplots()
plt.hist(httpapi_df['call_time'], bins=20)
plt.title('Histogram of HttpApi call times', fontsize=18)
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Frequency')
plt.show()
print("")

# Scatterplot
plt.title('Scatter plot HttpApi call times (ms)', fontsize=18)
sns.regplot(x="index", y="call_time", data=httpapi_df.reset_index(), ci=None)
plt.show()
```

```python
import seaborn as sns

# Side-by-side boxplots
plt.title('Call time distributions', fontsize=18)
sns.boxplot(x='type',y='call_time',data=call_times_df)
plt.show()
print("")
```
